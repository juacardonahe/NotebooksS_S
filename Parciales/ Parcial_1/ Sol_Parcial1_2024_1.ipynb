##Pregunta #1

La distancia media entre dos señales periódicas $x_1(t)\in\mathbb{R,}\ \mathbb{C}$ y $x_2(t)\in\mathbb{R,}\ \mathbb{C}$; se puede expresar a partir de la potencia media de la diferencia entre ellas:

$$
d^2(x_1, x_2) = \bar{P}_{x_1-x_2}=\lim_{T \to \infty} \frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}~dt.
$$

Sea $x_1(t)$ y $x_2(t$) dos señales periódicas de periodo $T\in\mathbb{R}^+,$ como se muestra a continuación:

$$
x_1(t) = Ae^{j\omega_0 t}$$

$$
x_2(t) = Be^{j5\omega_0 t}
$$

con $\omega_0 = \frac {2\pi}{T} ; \quad T, A, B \in \mathbb{R}^+$. **Determine la distancia entre las dos señales.**


---



###**Solución:**

Según la integral de potencia se tiene que:

$$\frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}dt = \frac{1}{T}\left[\int_T |x_1(t)|^2dt-2\int_T x_1(t)x_2(t)dt +\int_T|x_2(t)|^2dt\right] $$




Donde:
$$E_{x_1} = \int_T |x_1(t)|^2dt $$

$$E_{x_2} = \int_T |x_2(t)|^2dt $$

Por lo tanto la integral original se puede simplificar en terminos de potencias medias:

$$\bar{P}_{x_1}=\frac{1}{T}\int_T |x_1(t)|^2dt$$

$$\bar{P}_{x_2}=\frac{1}{T}\int_T |x_2(t)|^2dt$$

Quedando que:

$$\boxed{\frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}dt = \bar{P}_{x_1} -\frac{2}{T}\int_T x_1(t)x_2(t)dt + \bar{P}_{x_2}}$$

Ahora resolviendo para $\bar{P}_{x_1}$ se tiene que como $x_1(t) \in \mathbb{C}$, entonces $|x_1(t)|^2 = x_1(t) x_1(t)^*$

$$\bar{P}_{x_1}=\frac{1}{T}\int_T |x_1(t)|^2dt = \frac{1}{T} \int_{0}^{T} Ae^{j\omega_0 t} \cdot Ae^{-j\omega_0 t} dt = \frac{A^2}{T} \int_{0}^{T} e^{j\omega_0 t} \cdot e^{-j\omega_0 t} dt = \frac{A^2}{T} \int_{0}^{T} dt = \frac{A^2}{T} \cdot t\Big|_0^T = \frac{A^2}{T} [T - 0] $$

$$\boxed {\bar{P}_{x_1}= A^2}$$

Ahora resolviendo para $\bar{P}_{x_2}$ se tiene que como $x_2(t) \in \mathbb{C}$, entonces $|x_2(t)|^2 = x_2(t) x_2(t)^*$

$$\bar{P}_{x_2}=\frac{1}{T}\int_T |x_2(t)|^2dt = \frac{1}{T} \int_{0}^{T} Be^{j5\omega_0 t} \cdot Be^{-j5\omega_0 t} dt = \frac{B^2}{T} \int_{0}^{T} e^{j5\omega_0 t} \cdot e^{-j5\omega_0 t} dt = \frac{B^2}{T} \int_{0}^{T} dt = \frac{A^2}{T} \cdot t\Big|_0^T = \frac{B^2}{T} [T - 0] $$

$$\boxed {\bar{P}_{x_2}= B^2}$$

Para el otro termino se tiene que:

$$\frac{-2}{T} \int_{T} x_1(t) \cdot x_2(t) dt$$

$$ = \frac{-2}{T} \int_{0}^{T} Ae^{j\omega_0 t} \cdot Be^{j5\omega_0 t} dt$$

$$= \frac{-2}{T} \int_{0}^{T} Ae^{j\frac {2\pi}{T} t} \cdot Be^{j5\frac {2\pi}{T} t} dt $$

$$= \frac{-2}{T} \int_{0}^{T} A \cdot B \cdot e^{j\frac {12\pi}{T} t} dt $$

$$= \frac{-2}{T} A \cdot B \int_{0}^{T} e^{j\frac {12\pi}{T} t} dt $$

$$= \frac{-2}{T} A \cdot B \ [ \frac{T}{j12\pi} e^{j\frac {12\pi}{T} t}]\Big|_0^T $$

$$= \frac{-2}{T} A \cdot B \ [\frac{T}{j12\pi} e^{j\frac {12\pi}{T} T} - \frac{T}{j12\pi} e^{j\frac {12\pi}{T} 0} ]$$

Como $ e^{j{12\pi}} = cos(12\pi) + jsen(12\pi) = 1 + 0 = 1$ entonces esta expresion finalmente queda como:


$$\boxed {\frac{-2}{T} \int_{T} x_1(t) \cdot x_2(t) dt = \frac{-2}{T} A \cdot B \ [0] = 0}$$

Por lo tanto la integral original queda:

$$\boxed{\frac{1}{T} \int_T |x_1(t)-x_2(t)|^{2}dt = \bar{P}_{x_1} - 0 \ + \bar{P}_{x_2} = \bar{P}_{x_1} + \bar{P}_{x_2} = A^2 + B^2}$$

Para lo cual se obtiene:
$$d^2(x_1, x_2) = \lim_{T \to \infty} A^2 + B^2 $$

$$\boxed{d(x_1, x_2) = \sqrt {A^2 + B^2}} $$



---



###**Simulación distancia media:**

```python
import sympy as sp
from sympy import * #importar todo sympy
from sympy import exp

t = symbols('t', Real = True, nonzero = True, nonnegative = True)
T = symbols('T', Real = True, nonzero = True, nonnegative = True)
A = symbols('A', Real = True, nonzero = True, nonnegative = True)
B = symbols('B', Real = True, nonzero = True, nonnegative = True)

o1 = A*exp(sp.I*((2*pi)/T)*t)
o11 = sp.conjugate(o1)
o2 = B*exp(sp.I*((10*pi)/T)*t)
o22 = sp.conjugate(o2)

o0 =  2/T * integrate(o1*o2, (t, 0, T))
Pm1 = 1/T * integrate(o1*o11, (t, 0, T))
Pm2 = 1/T * integrate(o2*o22, (t, 0, T))
Pmf = limit(Pm1 + Pm2 + o0, T, oo)

distancia = Pmf

print('\nLa distancia media al cuadrado entre las dos señales es:')
distancia
```

## Pregunta #2

¿Cuál es la señal obtenida en tiempo discreto al utilizar un conversor análogo digital con frecuencia de muestreo de $5kHz$, aplicado a la señal continua $x(t) = 3 \cos(1000\pi t) + 5 \sin(2000\pi t) + 10 \cos(11000\pi t)$?. Realizar la simulación del proceso de discretización. En caso de que la discretización no sea apropiada, diseñe e implemente un conversor adecuado para la señal estudiada.


---

###**Solución:**

Se reemplaza $t$ en continuo por la discretización $t= \frac{n}{F_s}$, con $n\in\mathbb{Z}$, obteniendo asi lo siguiente:

$$x[t = \frac{n}{F_s}] = 3\cos[1000\pi\frac{n}{F_s}]+ 5\sin[2000\pi\frac{n}{F_s}] + 10\cos[11000\pi\frac{n}{F_s}],$$

Del enunciado se tiene la siguiente información:
$$F_s = 5000 \ [Hz]$$

Reemplazando este valor en la ecuación inicial se tiene que:

$$x[n] = 3\cos[1000\pi\frac{n}{5000}]+ 5\sin[2000\pi\frac{n}{5000}] + 10\cos[11000\pi\frac{n}{5000}],$$


Simplificando queda que:

$$x[n] = 3\cos[\frac{\pi}{5} \ n]+ 5\sin[\frac{2\pi}{5} \ n] + 10\cos[\frac{11\pi}{5} \ n],$$

Nombrando y ordenando las señales, para saber sus velocidades angulares discretas:

$x_1[n] = 3\cos[\frac{\pi}{5} \ n]$ $\rightarrow \Omega_1 = \frac{\pi}{5}$

$x_2[n] = 5\sin[\frac{2\pi}{5} \ n]$ $\rightarrow \Omega_2 = \frac{2\pi}{5}$

$x_3[n] = 10\cos[\frac{11\pi}{5} \ n]$ $\rightarrow \Omega_3 = \frac{11\pi}{5}$

Se observa que $\Omega_3$ es una frecuencia copia ya que $\Omega_3>\pi$, por lo tanto se debe encontrar la original.

Hallando la original de $x_3[n]$, se le resta $2\pi$ ya que $\Omega_3>\pi$

$$\boxed{\Omega' = \Omega_3 -2\pi = \frac{\pi}{5}}$$


Reemplazando este valor en la función discretizada:

$$x[n] = 3\cos[\frac{\pi}{5} \ n]+ 5\sin[\frac{2\pi}{5} \ n] + 10\cos[\frac{\pi}{5} \ n],$$

Por ende la señal discretizada obtenida con una frecuencia de muestreo $F_s = 5000 \ [Hz]$:

$$\boxed{x[n] = 13\cos[\frac{\pi}{5} \ n]+ 5\sin[\frac{2\pi}{5} \ n]}$$




Es evidente que la discretización propuesta no es apropiada ya que nos genera un ALIAS, por lo tanto haciendo uso de $Nyquist$ para hallar una frecuencia de muestreo $F_s$ apropiada, se llega a que:

$$\hat{F_s} \geq 2 \ F_{\text{Max}}$$

Definimos que nuestra $F_{\text{Max}}$ se relaciona con $\omega_3 = 11000 \pi$, para lo cual se tiene que:

$$F_{\text{Max}} = \frac{\omega_3}{2 \pi} = \frac{11000 \pi}{2 \pi} =  5500 \ [Hz] $$

Por lo tanto reemplazando en $Nyquist$ se obtiene lo siguiente:

$$\hat{F_s} \geq 2 (5500) \ [Hz]$$

$$\boxed{\hat{F_s} \geq 11000 \ [Hz]}$$


Ahora analizando la señal con $\hat{F_s}$, se llega a que:

$$x[n] = 3\cos[1000\pi\frac{n}{11000}]+ 5\sin[2000\pi\frac{n}{11000}] + 10\cos[11000\pi\frac{n}{11000}]$$

Simplificando queda que:

$$x[n] = 3\cos[\frac{\pi}{11} \ n]+ 5\sin[\frac{2\pi}{11} \ n] + 10\cos[\pi \ n]$$

Nombrando y ordenando las señales, para saber sus velocidades angulares discretas:

$x_1[n] = 3\cos[\frac{\pi}{11} \ n]$ $\rightarrow \Omega_1 = \frac{\pi}{11}$

$x_2[n] = 5\sin[\frac{2\pi}{11} \ n]$ $\rightarrow \Omega_2 = \frac{2\pi}{11}$

$x_3[n] = 10\cos[\pi \ n]$ $\rightarrow \Omega_3 = \pi $

Es evidente que haciendo uso de $Nyquist$ se corrige el problema del ALIAS, obteniendo asi la siguiente señal:

$$\boxed{x[n] = 3\cos[\frac{\pi}{11} \ n]+ 5\sin[\frac{2\pi}{11} \ n] + 10\cos[\pi \ n]}$$

---

###**Simulación discretización**

$x(t) = 3 \cos(1000\pi t) + 5 \sin(2000\pi t) + 10 \cos(11000\pi t)$

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la señal continua
def x(t):
    return 3 * np.cos(1000 * np.pi * t) + 5 * np.sin(2000 * np.pi * t) + 10 * np.cos(11000 * np.pi * t)

# Parámetros de la simulación
duration = 0.01  # Duración de la señal en segundos
sampling_rate = 5000  # Frecuencia de muestreo en Hz

# Intervalo de tiempo de la simulación
t = np.linspace(0, duration, int(duration * sampling_rate), endpoint=False)

# Muestreo de la señal continua
samples = x(t)

# Gráfica de la señal continua y los puntos muestreados
plt.plot(t, samples, label='Señal continua') #Señal continua
plt.stem(t, samples, markerfmt='ro', linefmt='r-', basefmt='k-', label='Puntos muestreados') #Señal discretizada

# Agregar líneas punteadas para cada funcion separada
plt.plot(t, 3 * np.cos(1000 * np.pi * t), '--', label='Velocidad angular 1000π rad/s')
plt.plot(t, 5 * np.sin(2000 * np.pi * t), '--', label='Velocidad angular 2000π rad/s')
plt.plot(t, 10 * np.cos(11000 * np.pi * t), '--', label='Velocidad angular 11000π rad/s')

plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.title('Aliasing con frecuencia de muestreo inapropiada')
plt.show()
```

Es evidente y se comprobó en la solución del enunciado que 5000 Hz nos genera un ALIAS y por lo tanto se pierde información, en la grafica anterior se evidencia como con esta frecuencia de muestreo no se alcanza a muestrear toda la señal.

##Pregunta #3

Implemente una simulación para encontrar la salida del sistema lineal e invariante al tiempo $\mathcal{H}\{\cdot\}$, con respuesta al escalón $h_\epsilon[n]=\{2,4,1,5,0,10\}$, $n\in\mathbb{Z}$, ante la entrada análoga en corriente $x(t)=20(\cos(t/3)+\cos(t/4)) [A]$. A: Amperios. Incluya los acondicionamientos necesarios de discretización y cuantización, asumiendo un microprocesador de 4 bits con entrada análoga de 4 mA a 20 mA.


---



###**Solución:**

Para la señal $x(t)$, se tiene que:

$\frac{\omega_1}{\omega_2} =\frac{1/3}{1/4} = \frac{4}{3} \in \mathbb{Q}$; entonces $x(t)$ es cuasiperiódica, con un período total:

$$T_1 = \frac{2\pi}{\omega_1} = \frac{2\pi}{1/3} = 6\pi$$

$$T_2 = \frac{2\pi}{\omega_2} = \frac{2\pi}{1/4} = 8\pi$$

El $MCM(6,8)=24,$ así:

$$\boxed{T = 24\pi \ [s]}$$

---

###**Simulación conversor y salida SLIT**
Entrada: $x(t)=20(\cos(t/3)+\cos(t/4))$

```python
import numpy as np # importar libreria numpy

def my_ceropen(x,ymin = 0.004, ymax = 0.02): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv

from scipy.spatial.distance import cdist

#Señal original

#Se defininen las frecuencias por separadas
F1 = (1/3)/(2*(np.pi))
F2 = (1/4)/(2*(np.pi))

#Se define cada periodo por separado
T1 = 1/F1
T2 = 1/F2

#Periodo total, se saca por medio del MCM de T1, T2
TT = 24*np.pi

#Aplicar y asegurar Nyquist para la Fs con la Fmax, en este caso F3
Fs = 10*max(F1,F2)

#Periodo de muestreo
Ts = 1/Fs

tv = np.arange(0,2*TT,Ts)#procesar dos periodos discretizados

y = 20*(np.cos(tv/3)+np.cos(tv/4))

#cero y pendiente
ycs = my_ceropen(y, ymin=0.004,ymax=0.02)

#número de bits y vector de cuantización
nbits = 4
rmin = 0.004
rmax = 0.02
ve = np.linspace(rmin,rmax,2**nbits) # (rmax-rmin)/(2^#bits - 1) -> si se quiere incluir rmin y rmax

#distancias y cero pendiente a vector cuantizacion
d = cdist(ycs.reshape(-1,1),ve.reshape(-1,1)) #se utiliza reshape(-1,1) para generar vectores columna como lo pide cdist
d.shape # d es una matriz de 500 x 8, ej: la fila 1 incluye la distancia euclidea entre y[0] y los 8 elementos en ve
```

```python
#se requiere identificar el elemento ve[j] más cercano a y[i] para genera señal cuantizada
ind = np.argmin(d,axis=1) #el parámetro axis = 1 indica que busca la posición a lo largo de las columnas del elemento más pequeño en cada fila
ind
```

```python
#función cuantización punto a punto

def my_cuantizador(yn, vq) : #yn punto a #cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamaño vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1))#distancia yn a vector estados, reshape(-1,1) asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posición distancia min
  return vq[ind]
```

```python
#recorrer vector cero y spam discretizado para cuantizar

for i in range(ycs.shape[0]):
   if i==0: #primera muestra
     yq = my_cuantizador(ycs[i],ve) #Funcion cuantizada almacena los valores cuantizados
   else:
     yq = np.append(yq,my_cuantizador(ycs[i],ve)) #apilar datos cuantizando si i>0
```

```python
#graficar señales

plt.plot(tv,ycs,c='r', label='ycs')#cs continua

plt.stem(tv,ycs, label='ycs_dis')#cs discreta

#estados de cuantización

for v in ve:
  plt.plot(tv,v*np.ones(len(tv)))

# graficar señal Digital (discreta + cuantizada)

plt.plot(tv,yq,c='y',marker='o',label='ydig', linewidth=2)

plt.legend()
plt.grid()
plt.xlabel('t')
plt.ylabel('Amplitud')
plt.show()
```

###**Determinación de la salida del SLIT**

Como ya se tiene la señal discretizada y cuantizada

Una vez que hemos discretizado y cuantificado la señal, procedemos a determinar la salida del sistema lineal e invariante en el tiempo (SLIT) mediante la convolución entre la entrada digital y la respuesta al impulso.

```python
he = np.array([2,4,1,5,0,10]) #respuesta al escalón del slit
h = np.diff(he)#respuesta al impulso del slit
print(h)
```

```python
w = np.convolve(yq,h,mode='same')#respuesta del slit -> se asegura tamaños iguales con mode=same
#graficar señal
plt.stem(tv,w,'b',label='$w[n]$')
plt.legend()
plt.xlabel('n [muestra]')
plt.ylabel('Amplitud')
plt.show()
```
